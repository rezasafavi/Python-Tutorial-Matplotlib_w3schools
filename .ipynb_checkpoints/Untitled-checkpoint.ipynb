```python
# Draw a line in a diagram from position (0,0) to position (6,250):
import matplotlib.pyplot as plt
import numpy as np 

xpoint = np.array([0,6])
ypoint = np.array([0,250])

plt.plot(xpoint , ypoint)
plt.show()
```

```python
# Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):
xpoint = ([1,8])
ypoint = ([3,10])

plt.plot(xpoint , ypoint , "o")
plt.show()
```

```python
# Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):
xpoint = ([1,2,6,8])
ypoint = ([3,8,1,10])

plt.plot(xpoint ,ypoint)
plt.show()
```

```python
#  Plotting without x-points :

ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint)
plt.show()
```

```python
# Mark each point with a circle:

ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint , marker = "o")
plt.show()
```

```python
# Use a dotted line:
ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint , linestyle = "dotted")
plt.show()
```

```python
# Use a dashed line:
ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint , linestyle = "dashed")
plt.show()
```

```python
# Set the line color to red:
ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint , ls = "dashed" , color = "r")
plt.show()
```

```python
#  Plot with a 20.5pt wide line:
ypoint = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoint , c = "g" , linewidth = '15.5')
plt.show()
```

```python
# Draw two lines by specifying a plt.plot() function for each line:
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1 , color = "pink")
plt.plot(y2 , color = "violet")

plt.show()
```

```python
# Draw two lines by specifiyng the x- and y-point values for both lines:
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
# Add labels to the x- and y-axis:
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330]) 

plt.plot(x,y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
# Add a plot title and labels for the x- and y-axis:
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
# Set font properties for the title and labels:

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
# Add grid lines to the plot:
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.grid()
plt.show()
```

```python
# Display only grid lines for the x-axis:
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.grid(axis= 'x')
plt.show()
```

```python
# Display only grid lines for the y-axis:
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.grid(axis= 'y')
plt.show()
```

```python
# Set the line properties of the grid:
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sport Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.grid(color ="red" , linestyle = '--' , linewidth = 0.5)
plt.show()
```
